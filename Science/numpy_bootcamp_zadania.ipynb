# Indeksowanie, wycinanie, iterowanie - Indexing, Slicing and Iterating

### Stwórz liste sześcianów liczb od 0 do 9

```python
x = np.arange(9)**3
x
```

### wybierz z listy zaznaczone elementy

[ 0, 1, **8**, **27**, **64**, 125, 216, 343, 512, 729]

    

```python
x[2:5]
```

[ **0**, 1, **8**, 27, **64**, 125, 216, 343, 512, 729]

```python
x[:5:2]
```

### przekształć by otrzymać wynik

[-1000, 1,  -1000, 27, -1000,   125,   216,   343,   512, 729]

```python
x[:5:2] = -1000
x
```

### Odwróć listę

```python
x[::-1]
```

### Sprawdź typ 

```python
x.dtype
```

### W pętli for wyprintuj dla każdego elementu pierwiastek sześcienny z elementy

```python
for i in x:
   print(i**(1/3))
x ** (1/3)    
    
```

### Stwórz macierz 10x10 z kwadratami liczb od 1 do 100

```python
k = np.arange(1, 101)**2
k = k.reshape((10,10))
print(k)
```

### Stwórz macierz 5x5 gdzie na krawędziach będą 0 a w środku 1

```python
np.zeros((3,3))  #tablica wypełniona zerami
```

```python
np.ones((2,2))
```

```python
x = np.zeros((5,5))
x[:2,-3:] = 10     #do drugiego trzy ostatnie elementy równają się 10, reszta 0
x
```

```python
x = np.zeros((5,5))
#  x[1:4,1:4] = 1        to zadziała
x[1:-1,1:-1] = 1    #to będzie niezależne o wielkości macierzy
x
```

### Stwórz tabliczkę mnożenia liczb od 0 do 9 w postaci dwuwymiarowej macierzy

```python
np.arange(5) * np.array([2,3,4,5,6]).reshape(5,1)  #resh na 5 wierszy i 1 kolumna --> transpozycja i mnożenie macierzy
```

```python
np.arange(1,11) * np.arange(1,11).reshape(10,1)
```
